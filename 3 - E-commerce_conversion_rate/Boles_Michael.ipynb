## Initial Data Analysis

### Loading packages and data set

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Import data set
data_raw = pd.read_csv('./conversion_data.csv')
```

### Checking size of data set and data types

```python
# check shape of raw data table and data type across columns
print(data_raw.shape)
print(data_raw.dtypes)
```

### Looking at most common values, frequency of missing values

```python
# get most common values
values = pd.DataFrame()
for column in data_raw:
    value_counts = data_raw[column].value_counts() 
    values[str(column)] = pd.Series(value_counts.index[:10])
values
```

```python
# for those most common values, get corresponding counts
counts = pd.DataFrame()
for column in data_raw:
    value_counts = data_raw[column].value_counts()
    counts[str(column)] = pd.Series(value_counts.values[:10])
counts
```

```python
# check number of missing values across columns 
columns = list(data_raw.columns)
percent_missing = [round(data_raw[column].isna().mean()*100,2) for column in columns]
missing_dict = dict(zip(columns, percent_missing))
missing_dict
```

## Data Wrangling

### Duplicates

  - This dataset has 95% duplicate entries
  - But without unique user ID, I have to assume that each observation may be a unique user/event 

```python
# check for duplicate entries
dups = data_raw[data_raw.duplicated(keep='first')]
print('Duplicate fraction is: ' + str(100*round(len(dups)/len(data_raw),3)) + '%')
```

### Outlier removal

 - Four columns are categorical (country, new_user, source, converted) -- no outlier removal necessary
 - Remaining two (age, total_pages_visited) are numeric -- will not remove outliers here 

## Exploratory Data Analysis

### Data summary

```python
# Examine data summary
data_raw.describe()
```

### Feature histograms 

```python
# Plot distributions of numerical features

data_numerical = data_raw.drop(columns=['country','source'])

f, axes = plt.subplots(1, 4, figsize=(16,4))
for i, ax in enumerate(axes.flatten()):

    # get feature data
    data = data_numerical.iloc[:, i]
    
    # set up textbox
    average = round(np.nanmean(data), 2)
    median = round(np.nanmedian(data), 2)
    stdev = round(np.std(data), 2)
    props = dict(facecolor='white', edgecolor='none', alpha=0.67, boxstyle='square, pad=1')
    textbox = str(data_numerical.columns[i]) + ' \nAverage = %.1f \nMedian = %.1f \nStdev = %.1f' % (average, median, stdev)
    
    try:
        sns.distplot(data, ax=ax)
        ax.text(0.55, 0.75, textbox, fontsize=12, transform=ax.transAxes)
    except:
        sns.countplot(data, ax=ax)
        ax.text(0.55, 0.75, textbox, fontsize=12, transform=ax.transAxes)
    
plt.tight_layout()
```

Feature histogram conclusions
 - Average age of visitors is 30 years
 - There are twice as many new users as returning users
 - Most common number of page views is 4 per session
 - Non-conversions vastly outnumber conversions (by ~10x) 

### Conversion analysis

```python
# separate user information by conversion status
data_converted = data_raw[data_raw['converted'] == 1]
data_not_converted = data_raw[data_raw['converted'] == 0]
print(data_raw.columns)
fract_converted = len(data_converted)/(len(data_converted)+len(data_not_converted))
print('Overall conversion fraction: %s' % str(100*fract_converted))
```

```python
# create dist plots for page visits and age

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (16,8))

# plot conversion differences versus total pages visited
sns.distplot(data_converted['total_pages_visited'], label='Converted', ax=ax1)
sns.distplot(data_not_converted['total_pages_visited'], label='Not Converted', ax=ax1, kde_kws={'bw':1})
ax1.set_xlim([0, 30])
ax1.legend(fontsize=24)
ax1.get_yaxis().set_visible(False)
ax1.set_xlabel('Total pages visited', fontsize=24)
ax1.tick_params(axis='both', which='major', labelsize=24)

# plot conversion differences versus age
sns.distplot(data_converted['age'], label='Converted', ax=ax2)
sns.distplot(data_not_converted['age'], label='Not Converted', ax=ax2)
ax2.set_xlim([16, 60])
ax2.legend(fontsize=24)
ax2.get_yaxis().set_visible(False)
ax2.set_xlabel('Age', fontsize=24)
ax2.tick_params(axis='both', which='major', labelsize=24)

plt.tight_layout()
plt.show()
```

Conversion comparison vs. page visits and age
 - Converting customers visit 15 pages on average, vs. 3 for non-converting visitors
 - May be unsurprising that engaged users convert
 - Converting customers are 30 years old on average, while non-converting are 22 years old
 - Younger visitors may have less disposable income to put towards purchase

```python
# create plots: converted vs. not converted for other 5 features

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,5))
fontsize = 20

# plot fraction converted by new user status 
conv_freq_by_new_status = 100*data_converted['new_user'].value_counts()/(data_converted['new_user'].value_counts()+data_not_converted['new_user'].value_counts())
g = sns.barplot(x=conv_freq_by_new_status.index, y=conv_freq_by_new_status.values, ax=ax1)
ax1.set_ylabel('Converted (%)', fontsize=fontsize)
ax1.set_xlabel('User status', fontsize=fontsize)
ax1.tick_params(axis='both', which='major', labelsize=fontsize)
g.set(xticklabels=['Old', 'New'])

# plot fraction converted by source 
conv_freq_by_new_status = 100*data_converted['source'].value_counts()/(data_converted['source'].value_counts()+data_not_converted['source'].value_counts())
sns.barplot(x=conv_freq_by_new_status.index, y=conv_freq_by_new_status.values, ax=ax2)
ax2.set_ylabel('Converted (%)', fontsize=fontsize)
ax2.set_xlabel('Channel', fontsize=fontsize)
ax2.tick_params(axis='both', which='major', labelsize=fontsize)

plt.subplots_adjust(wspace=.33)
# plt.tight_layout()

plt.show()
```

Conversion comparison vs. user status and channel
 - Users who have previously visited web page are more likely to convert
 - May be another example of semi-obvious observation that engagement predicts conversion
 - No strong correlation between conversion success and channel (search engine, ad click, or direct html entry)

```python
# create plots: converted vs. not converted for other 5 features

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,5))
fontsize = 20

# plot fraction converted by country 
order = ['US', 'China', 'UK', 'Germany']
sns.barplot(x=data_raw['country'].value_counts().index, 
            y=data_raw['country'].value_counts().values/1000, 
            order=order,
            ax=ax1)
ax1.set_ylabel('Users (thousands)', fontsize=fontsize)
ax1.set_xlabel('Country', fontsize=fontsize)
ax1.tick_params(axis='both', which='major', labelsize=fontsize)
ax1.tick_params(labelrotation=45)


# plot fraction converted by country 
conv_freq_by_country = 100*data_converted['country'].value_counts()/(data_converted['country'].value_counts()+data_not_converted['country'].value_counts())
sns.barplot(x=conv_freq_by_country.index, 
            y=conv_freq_by_country.values, 
            order=order,
            ax=ax2)
ax2.set_ylabel('Converted (%)', fontsize=fontsize)
ax2.set_xlabel('Country', fontsize=fontsize)
ax2.tick_params(axis='both', which='major', labelsize=fontsize)
ax2.tick_params(labelrotation=45)

plt.subplots_adjust(wspace=.33)
# plt.tight_layout()
plt.show()
```

User counts and conversion vs. country
 - Highest conversion in countries with fewest users (Germany, UK)
 - China accounts for 25% of user base but just 1% of conversions!
 - Should evaluate source of poor conversion in China

### Calculations

**Conversion frequency** varies significantly across geography
 - Europe: 5 - 6% conversion
 - US: 4% conversion
 - **China: 0.1% conversion (!)**

```python
# display conversion frequency (%) across countries
conv_freq_by_country
```

**China** is under-represented among conversions
- Chinese are 24% of user base
- But are just 1% among conversions
- Improving conversions in Chinese market may be low-hanging fruit

```python
# Display China statistics
print('Users:\n' + str(data_raw['country'].value_counts()) + '\n')
print('China user fraction:\n' + str(round(data_raw['country'].value_counts()[1]/data_raw['country'].value_counts().sum(),2)) + '\n')
print('Conversions:\n' + str(data_converted['country'].value_counts()) + '\n')
print('China conversion fraction:\n' + str(data_converted['country'].value_counts()[3]/data_converted['country'].value_counts().sum()))
```

**Action in China** should be prioritized
 - Western users on average convert in 5% of sessions
 - If Chinese consumers converted at 5%, we might expect
  - 3900 additional conversions
  - 38% boost in total conversions

```python
# display average conversion among Germany, UK, US
avg_conv_west = conv_freq_by_country[1:].values.sum()/3
print('Conversion average in the West: ' + str(round(avg_conv_west,2)) + '%\n')

# calculate number of China conversions if conversion rate were average in the West  
china_users = data_raw['country'].value_counts()['China']
china_converts_with_5percent = china_users*avg_conv_west/100
print('With 5.1% conversion, number of Chinese users converting: ' + str(int(china_converts_with_5percent)) + '\n')

# calculate boost to overall number of conversions in this case
total_conversions_boost = ( len(data_converted) + int(china_converts_with_5percent) ) / len(data_converted)
print('With 5.1% conversion in China, boost to overall conversions: ' + str(round(total_conversions_boost,2)) + 'x\n')
```
